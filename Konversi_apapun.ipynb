```python

import os
from PIL import Image
from docx import Document
from pptx import Presentation
from PyPDF2 import PdfReader, PdfWriter, PdfMerger
import pandas as pd
from reportlab.pdfgen import canvas
from pydub import AudioSegment
from google.colab import files

# Utility Functions
def upload_file():
    """Upload file and return its path."""
    uploaded = files.upload()
    for name in uploaded.keys():
        return name

def download_file(file_path):
    """Download file."""
    files.download(file_path)

# **Image Conversion Features**
def convert_png_to_jpeg(input_file):
    img = Image.open(input_file)
    output_file = input_file.rsplit('.', 1)[0] + '.jpg'
    img.convert('RGB').save(output_file, 'JPEG')
    return output_file

def convert_jpeg_to_png(input_file):
    img = Image.open(input_file)
    output_file = input_file.rsplit('.', 1)[0] + '.png'
    img.save(output_file, 'PNG')
    return output_file

def compress_image(input_file):
    img = Image.open(input_file)
    output_file = input_file.rsplit('.', 1)[0] + '_compressed.jpg'
    img.save(output_file, 'JPEG', optimize=True, quality=20)
    return output_file

# **Document Conversion Features**
def convert_docx_to_pdf(input_file):
    doc = Document(input_file)
    output_file = input_file.rsplit('.', 1)[0] + '.pdf'
    pdf = canvas.Canvas(output_file)
    for para in doc.paragraphs:
        pdf.drawString(100, 800, para.text)
    pdf.save()
    return output_file

def convert_pdf_to_docx(input_file):
    pdf = PdfReader(input_file)
    output_file = input_file.rsplit('.', 1)[0] + '.docx'
    doc = Document()
    for page in pdf.pages:
        doc.add_paragraph(page.extract_text())
    doc.save(output_file)
    return output_file

def merge_docx(files_list):
    merged_doc = Document()
    for file in files_list:
        doc = Document(file)
        for para in doc.paragraphs:
            merged_doc.add_paragraph(para.text)
    output_file = "merged_document.docx"
    merged_doc.save(output_file)
    return output_file

# **PDF Features**
def compress_pdf(input_file):
    pdf = PdfReader(input_file)
    writer = PdfWriter()
    output_file = input_file.rsplit('.', 1)[0] + '_compressed.pdf'
    for page in pdf.pages:
        writer.add_page(page)
    writer.add_metadata({'/Author': 'Compressed'})
    with open(output_file, 'wb') as f:
        writer.write(f)
    return output_file

def merge_pdfs(file_list):
    merger = PdfMerger()
    for file in file_list:
        merger.append(file)
    output_file = "merged.pdf"
    merger.write(output_file)
    merger.close()
    return output_file

def split_pdf(input_file, start, end):
    pdf = PdfReader(input_file)
    writer = PdfWriter()
    output_file = f"split_{start}_{end}.pdf"
    for page in pdf.pages[start:end]:
        writer.add_page(page)
    with open(output_file, 'wb') as f:
        writer.write(f)
    return output_file

def encrypt_pdf(input_file, password):
    pdf = PdfReader(input_file)
    writer = PdfWriter()
    output_file = input_file.rsplit('.', 1)[0] + '_encrypted.pdf'
    for page in pdf.pages:
        writer.add_page(page)
    writer.encrypt(password)
    with open(output_file, 'wb') as f:
        writer.write(f)
    return output_file

# **Spreadsheet Features**
def convert_excel_to_pdf(input_file):
    df = pd.read_excel(input_file)
    output_file = input_file.rsplit('.', 1)[0] + '.pdf'
    df.to_html("temp.html")
    pdf = canvas.Canvas(output_file)
    with open("temp.html", 'r') as f:
        lines = f.readlines()
        for i, line in enumerate(lines):
            pdf.drawString(50, 800 - (i * 20), line.strip())
    pdf.save()
    os.remove("temp.html")
    return output_file

def convert_csv_to_excel(input_file):
    df = pd.read_csv(input_file)
    output_file = input_file.rsplit('.', 1)[0] + '.xlsx'
    df.to_excel(output_file, index=False)
    return output_file

def convert_excel_to_csv(input_file):
    df = pd.read_excel(input_file)
    output_file = input_file.rsplit('.', 1)[0] + '.csv'
    df.to_csv(output_file, index=False)
    return output_file

# **Presentation Features**
def convert_pptx_to_pdf(input_file):
    presentation = Presentation(input_file)
    output_file = input_file.rsplit('.', 1)[0] + '.pdf'
    pdf = canvas.Canvas(output_file)
    for i, slide in enumerate(presentation.slides):
        for shape in slide.shapes:
            if shape.has_text_frame:
                pdf.drawString(100, 800 - (i * 20), shape.text)
    pdf.save()
    return output_file

def convert_pdf_to_pptx(input_file):
    pdf = PdfReader(input_file)
    output_file = input_file.rsplit('.', 1)[0] + '.pptx'
    presentation = Presentation()
    for page in pdf.pages:
        slide = presentation.slides.add_slide(presentation.slide_layouts[5])
        textbox = slide.shapes.add_textbox(left=0, top=0, width=500, height=500)
        textbox.text = page.extract_text()
    presentation.save(output_file)
    return output_file

# **Audio/Video Features**
def convert_mp4_to_mp3(input_file):
    output_file = input_file.rsplit('.', 1)[0] + '.mp3'
    audio = AudioSegment.from_file(input_file, format="mp4")
    audio.export(output_file, format="mp3")
    return output_file

def convert_mp3_to_wav(input_file):
    output_file = input_file.rsplit('.', 1)[0] + '.wav'
    audio = AudioSegment.from_file(input_file, format="mp3")
    audio.export(output_file, format="wav")
    return output_file

# **Main Execution**
if __name__ == "__main__":
    print("Upload a file to start:")
    input_file = upload_file()
    print("Select an option:")
    options = [
        "1. Convert PNG to JPEG",
        "2. Convert JPEG to PNG",
        "3. Compress Image",
        "4. Convert DOCX to PDF",
        "5. Convert PDF to DOCX",
        "6. Compress PDF",
        "7. Merge PDFs",
        "8. Split PDF",
        "9. Encrypt PDF",
        "10. Convert Excel to PDF",
        "11. Convert CSV to Excel",
        "12. Convert Excel to CSV",
        "13. Convert PPTX to PDF",
        "14. Convert PDF to PPTX",
        "15. Convert MP4 to MP3",
        "16. Convert MP3 to WAV"
    ]
    for opt in options:
        print(opt)
    choice = int(input("Enter your choice: "))

    # Map choices to functions
    if choice == 1:
        output_file = convert_png_to_jpeg(input_file)
    elif choice == 2:
        output_file = convert_jpeg_to_png(input_file)
    elif choice == 3:
        output_file = compress_image(input_file)
    elif choice == 4:
        output_file = convert_docx_to_pdf(input_file)
    elif choice == 5:
        output_file = convert_pdf_to_docx(input_file)
    elif choice == 6:
        output_file = compress_pdf(input_file)
    elif choice == 7:
        file_list = [upload_file() for _ in range(2)]
        output_file = merge_pdfs(file_list)
    elif choice == 8:
        start, end = map(int, input("Enter start and end pages: ").split())
        output_file = split_pdf(input_file, start, end)
    elif choice == 9:
        password = input("Enter password: ")
        output_file = encrypt_pdf(input_file, password)
    elif choice == 10:
        output_file = convert_excel_to_pdf(input_file)
    elif choice == 11:
        output_file = convert_csv_to_excel(input_file)
    elif choice == 12:
        output_file = convert_excel_to_csv(input_file)
    elif choice == 13:
        output_file = convert_pptx_to_pdf(input_file)
    elif choice == 14:
        output_file = convert_pdf_to_pptx(input_file)
    elif choice == 15:
        output_file = convert_mp4_to_mp3(input_file)
    elif choice == 16:
        output_file = convert_mp3_to_wav(input_file)
    else:
        print("Invalid choice!")
        output_file = None

    if output_file:
        download_file(output_file)
```
